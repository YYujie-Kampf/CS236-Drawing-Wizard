```python
import os
import shutil

def move_and_delete_folder(source_folder, destination_folder, delete_folder, image_name='fake_B'):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Loop through files in the source folder
    for filename in os.listdir(source_folder):
        if image_name in filename:
            source_path = os.path.join(source_folder, filename)
            destination_path = os.path.join(destination_folder, filename)

            # Use shutil.move() to move the file
            shutil.move(source_path, destination_path)

    # Remove the original folder
    shutil.rmtree(delete_folder)
    print(f"Deleted original folder: {delete_folder}")
```

```python
source_directory = "/content/drive/Shareddrives/CS236_Drawing_Wizard/results/edges2handbags_pretrained/test_latest/images"
destination_directory = "/content/drive/Shareddrives/CS236_Drawing_Wizard/results/edges2handbags"
delete_directory = "/content/drive/Shareddrives/CS236_Drawing_Wizard/results/edges2handbags_pretrained"

move_and_delete_folder(source_directory, destination_directory, delete_directory, image_name='fake')
```
